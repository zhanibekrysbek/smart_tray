### Draw Tray Markers in Grids

```python
ind = 0
ids = [20,21,22,23,24,25,26,27]
gmlen = 0.113
gsep = 0.022

imboards = []
aruco_dict = aruco.Dictionary_get(aruco.DICT_4X4_50)

fig = plt.figure(figsize=(20,7))


for i in range(len(ids)/2):
    gboard = aruco.GridBoard_create(1,2,gmlen,gsep, aruco_dict, ids[i*2])
    imboard = gboard.draw((1000,2000))
    cv2.imwrite(
        os.path.join(base_dir, "tray_3d_board_ids_{}_{}.png".format(ids[i*2], ids[i*2+1])), imboard)

    imboards.append(imboard)
    
    plt.subplot(1,4,i+1)
    showMarker(imboard)
    plt.title('ID: {}, {}'.format(ids[i*2], ids[i*2+1]))

plt.show()
```

## 3D Tray Board

```python
mrklen = 0.115
ids = [20,21,22,23,24,25,26,27]

side_marker0_right = np.array(
             [[0., 0., -mrklen],
              [0., 0., 0.], #1
              [mrklen, 0., 0.],
              [mrklen, 0., -mrklen]], dtype=np.float32)

side_marker0_left = np.array(
             [[0., 0., -mrklen],
              [0., 0., 0.], #1
              [-mrklen, 0., 0.],
              [-mrklen, 0., -mrklen]], dtype=np.float32)

top_marker0 = np.array(
             [[0., 0., 0.],
              [mrklen, 0., 0.],
              [mrklen, -mrklen, 0.],
              [0., -mrklen, 0.]], dtype=np.float32)

# [x_off,y_off,z_off]
offsets_right = np.array(
                   [[-0.1930, -0.15796, -0.025],  # 20
                    [-0.05514,-0.15796, -0.025],  # 21
                    [0.0871,  -0.15796, -0.025],  # 22
                   ], dtype=np.float32) 

# [x_off,y_off,z_off]
offsets_left = np.array(
                   [[0.1931, 0.15796, -0.02],   # 24
                    [0.053,  0.15796, -0.02],   # 25
                    [-0.079,  0.15796, -0.02],  # 23
                   ], dtype=np.float32) 


# [x_off,y_off,z_off]
offsets_top = np.array(
                   [[0.07623, 0.1252, 0.],  # 26
                    [0.07623, -0.0129, 0.]  # 27
                   ], dtype=np.float32) 


board_corners = [side_marker0_right + offset for offset in offsets_right]
board_corners += [side_marker0_left + offset for offset in offsets_left]
board_corners += [top_marker0 + offset for offset in offsets_top]

board_ids = np.array( [[20],[21], [22], [24], [25], [23], [26], [27]], dtype=np.int32)
board = aruco.Board_create(board_corners, aruco_dict, board_ids)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
col = ['blue', 'red', 'black', 'yellow', 'pink', 'magenta', 'gray', 'blue']
for ind in range(len(board_corners)):
    x = np.array(board_corners[ind][:,0]).flatten()
    y = np.array(board_corners[ind][:,1]).flatten()
    z = np.array(board_corners[ind][:,2]).flatten()
    ax.scatter(x,y,z, c = col[ind])
```
