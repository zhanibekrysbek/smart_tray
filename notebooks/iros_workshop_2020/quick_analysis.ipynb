```python
def get_fxy(f, keyword = 'force', beta=0.9):
    eps = 1e-6
    f_xy =  exp_moving_average(np.sqrt(f[keyword][:,0]**2 + f[keyword][:,1]**2), beta = beta)
    # Fx/Fy
    prop_xy = np.arctan(f[keyword][:,0] / (f[keyword][:,1] + eps) )/np.pi*180
    prop_xy = exp_moving_average(prop_xy, beta=beta)
    # when the force is negligble assign the direction to zero
    small_idx = f_xy<0.5
    prop_xy[small_idx] = 0.0
    
    return f_xy, prop_xy
    
    
def get_axy(acc, beta=0.9):
    eps = 1e-6
    a_xy =  exp_moving_average(np.sqrt(acc['accel'][:,0]**2 + acc['accel'][:,1]**2), beta = beta)

    # Fx/Fy
    prop_xy = np.arctan(acc['accel'][:,0] / (acc['accel'][:,1] + eps) )/np.pi*180
    prop_xy = exp_moving_average(prop_xy, beta=beta)
    
    # when the force is negligble assign the direction to zero
    small_idx = a_xy<0.1
    prop_xy[small_idx] = 0.0
    
    return a_xy, prop_xy


def interpolate(fv, tv, N=1001):
    tv = tv.copy()/tv.max()
    tvv = np.linspace(0,1,1001)
    fvv = np.interp(tvv, tv, fv)
    
    return tvv, fvv

def realign(f, thresh = 5.):
    idx = f['force'][:,2]>thresh
    f['force'] = f['force'][idx,:]
    f['torque'] = f['torque'][idx,:]
    f['time_steps'] = f['time_steps'][idx]
    f['time_steps']-=f['time_steps'][0]
    return f

def integrate_(f, t):
    
    integ = integrate.cumtrapz(f, t)
    integ = np.append(integ, integ[-1])
    return integ
```

```python
eps = 1e-4
for obs in observations:
    
    for i in range(1,3):
        f = obs['rft'+str(i)]
        f = realign(f)        
        
        t_int, fint0 = interpolate(f['force'][:,0], f['time_steps'])
        _, fint1 = interpolate(f['force'][:,1], f['time_steps'])
        _, fint2 = interpolate(f['force'][:,2], f['time_steps'])
        f['force'] = np.vstack([fint0, fint1, fint2]).transpose()
        
        t_int, fint0 = interpolate(f['torque'][:,0], f['time_steps'])
        _, fint1 = interpolate(f['torque'][:,1], f['time_steps'])
        _, fint2 = interpolate(f['torque'][:,2], f['time_steps'])
        f['torque'] = np.vstack([fint0, fint1, fint2]).transpose()
        
        f['time_steps'] = t_int
        
        fint0 = integrate_(f['force'][:,0], t_int)
        fint1 = integrate_(f['force'][:,0], t_int)
        fint2 = integrate_(f['force'][:,0], t_int)
        f['force_integ'] = np.vstack([fint0, fint1, fint2]).transpose()
        
        f['f_xy'], f['prop_xy'] = get_fxy(f)
        
    obs['Fsum'] = obs['rft1']['force']+obs['rft2']['force']
    obs['Fsum_xy'], obs['Fsum_prop'] = get_fxy(obs,'Fsum', beta=0.5)
    acc = obs['imu']
    acc['accel'][:,0] = exp_moving_average(acc['accel'][:,0])
    acc['accel'][:,1] = exp_moving_average(acc['accel'][:,1])
    acc['accel'][:,2] = exp_moving_average(acc['accel'][:,2])

    acc['mag'][:,0] = exp_moving_average(acc['mag'][:,0])
    acc['mag'][:,1] = exp_moving_average(acc['mag'][:,1])
    acc['mag'][:,2] = exp_moving_average(acc['mag'][:,2])
    
    acc['gyro'][:,0] = exp_moving_average(acc['gyro'][:,0])
    acc['gyro'][:,1] = exp_moving_average(acc['gyro'][:,1])
    acc['gyro'][:,2] = exp_moving_average(acc['gyro'][:,2])
    
    acc['a_xy'], acc['prop_xy'] = get_axy(acc)
    acc['t_int'], acc['a_xy_int'] = interpolate(acc['a_xy'], acc['time_steps'])
```
